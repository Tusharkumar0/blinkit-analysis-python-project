## BUSINESS REQUIREMENTS

### **KPI's REQUIREMENTS**

```python
# TOTAL SALES
total_sales = df["Sales"].sum()
# AVERAGE SALES
average_sales = df["Sales"].mean()
# NO. OF ITEM SOLD
no_of_item_sold = df["Sales"].count()
print(df["Item Fat Content"].unique())

# display
print('Total sales =',total_sales)
print('Average sales =',average_sales)
print('No. of item sold =',no_of_item_sold)
```

### **analyze total sales for each item category**


```python
#analyze total sales for each item category

sales_by_item = df.groupby('Item Type')['Sales'].sum()
plt.figure(figsize=(10,6))
sales_by_item.plot(kind='bar')
plt.title('Total sales by item type')
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt .xticks(rotation=90)
for i,value in enumerate(sales_by_item):
    plt.text(i,value,f'{value:.0f}',
             ha='center',va='bottom',fontsize=8)
plt.show()
```

### **Total sales by outlet size**

```python
#analyze total sales by outlet size(bar chart)

sales_by_outlet = df.groupby('Outlet Size')['Sales'].sum()
plt.figure(figsize=(10,5))
sales_by_outlet.plot(kind='bar')
plt.title('Sales by outlet size')
plt.xlabel('Outlet Size')
plt.ylabel('Sales')
for i,value in enumerate(sales_by_outlet):
    plt.text(i,value,f'{value:.0f}',
             ha='center',va='bottom',fontsize=8)
plt.show()
```

### **Total sales by item fat content**

```python
sales_by_fat_content = df.groupby('Item Fat Content')['Sales'].sum()
plt.figure(figsize=(10,5))
plt.pie(sales_by_fat_content,
        labels=sales_by_fat_content.index,
        autopct='%1.1f%%',
        startangle=90)
plt.title('Sales by fat content')
plt.axis('equal')
plt.show()
```

### **Total sales by outlet location type**

```python
sales_by_outlet_location_type = df.groupby('Outlet Location Type')['Sales'].sum()
plt.figure(figsize=(5,6))
sales_by_outlet_location_type.plot(kind='bar')
plt.title("sales by location type")
plt.xlabel('Outlet Location Type')
plt.ylabel('Sales')
for i, value in enumerate(sales_by_outlet_location_type):
    plt.text(i,value,f'{value:.0f}',
             ha='center',va='bottom',fontsize=8)

plt.show()
           
```

### **rating vs sales**

```python
plt.figure(figsize=(5,6))
plt.scatter(df['Rating'],df['Sales'])

plt.title('Rating vs Sales')
plt.xlabel('Rating')
plt.ylabel('Sales')

plt.show()
```

###  **BLINKIT SALES DATA - BUSINESS INSIGHTS (POINTS):**

1) Total Sales: ₹12,01,681.48 | Avg Sale/Item: ₹140.99 | Items Sold: 8,523

2) Item Type: Fruits & Vegetables (₹1.78L) and Snack Foods (₹1.75L) are the top performers. Seafood (₹9K) and Breakfast (₹15.6K) are the weakest — inventory/promotion review is needed.
    
3) Outlet Size: Medium outlets generate the highest sales (₹5.08L), while High-size outlets generate the lowest (₹2.49L) — despite having more space, they are underperforming.
    
4) Fat Content: Low Fat items hold a 64.6% sales share vs Regular at 35.4% — indicating a health-conscious buying trend.
    
5) Location Type: Tier 3 cities are at the top (₹4.72L), followed by Tier 2 (₹3.93L), and then Tier 1 (₹3.36L) — emerging cities are performing better than metro areas.
    
6) Rating vs Sales: There is no clear correlation — high-rated items do not necessarily sell more. Rating is not driving the purchase decision.
    
7) Quick takeaway: Increase focus on Low Fat + top categories, replicate the Medium-outlet model, and continue expansion in Tier 2/3 cities.
